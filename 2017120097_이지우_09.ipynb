## 문제1. 적분 시각화

```python
import matplotlib.pyplot as plt
import numpy as np

def xaxis(x):
    return x*0

def ri_sum(func, xrange, resolution):
    plt.figure()
    a, b = xrange
    ori_x= np.linspace(a,b,10000)
    plt.plot(ori_x, func(ori_x),color='r')
    plt.plot(ori_x,xaxis(ori_x),color='k')
    
    sum = 0 # 리만합을 저장할 변수
    
    for i in range(resolution):  # resolution 만큼 bar를 만듭니다
        Width = (b-a)/resolution  # bar의 너비
        x = a + (Width*((i*2)+1)/2)  # 각 bar의 중간지점
        plt.bar(x,func(x),width = Width, color = 'blue')  #bar 출력
        
        area = Width*func(x) # 각 bar의 넓이를 너비와 중간지점의 함숫값으로 계산합니다
        sum += area
        
    
    plt.xlim(xrange)
    plt.show()
    
    result = sum
    print("함수의 리만합은: ", result)
    
    return sum
    
```

```python
def func(x):
    return (x)**2-10
ri_sum(func, (-10, 10), 50)
```

## 문제2. 비교해보기

실제로 치환적분한 값: 26/3

2x+1 = t로 치환해서 계산

t의 범위는 1 ~ 9가 되고, (1/3) X (t)^(3/2)에 대입해서 구했다.

근사화 해보기:

```python
def func(x):
    return np.sqrt(1+2*x)

real = 26/3 #계산한 값 ( 정답 )
error = 100 
resolution = 100 

while error > 1e-6 : #error 가 1e-6 보다 작을 때 까지 resolution을 늘리면서 실제값과 비교를 반복합니다.
    resolution = round(resolution *1.1)
    error = abs(real - ri_sum(func,(0,4),resolution)) # 과제1번에서 만든 함수에 return을 추가해서 사용했습니다.
    print('resolution = ', resolution)
    print('error = ', error)
    
```

resolution이 678 이상이면 err < 1e-6 을 만족한다.
